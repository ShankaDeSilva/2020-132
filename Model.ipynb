```python
import pandas as pd
import re
dataset = pd.read_csv('Train_Dataset.csv')
```

```python
dataset
```

```python
data = dataset.replace('[^\d.]', '', regex=True).astype(float)
data
```

```python
data.dtypes
```

```python
features = data.iloc[:,0:6]

features.head()
```

```python
labels = data.Labels
labels.head()
```

```python
from sklearn.model_selection import train_test_split

train_data,test_data,train_target,test_target=train_test_split(features,labels,test_size = 0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier

algorithm = KNeighborsClassifier()

algorithm.fit(train_data,train_target)
```

```python
train_data.shape
```

```python
test_data.shape
```

```python
predicted_targets = algorithm.predict(test_data)
```

```python
print('Actual Targets:',test_target)
print('Predicted Targets:',predicted_targets)
```

```python
from sklearn.metrics import accuracy_score

acc=accuracy_score(test_target,predicted_targets)
print('Accuracy:', acc)
```
